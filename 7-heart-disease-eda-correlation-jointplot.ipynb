```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings


warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
df = pd.read_csv("/kaggle/input/heart-data/heart_data.csv")
```

```python
features_to_compare = [('Age', 'MaxHR'), ('Oldpeak', 'MaxHR'), ('Cholesterol', 'Age')]
for x, y in features_to_compare:
    sns.jointplot(data=df, x=x, y=y, hue="HeartDisease", kind="scatter", palette='Set1')
    plt.suptitle(f"Joint Plot: {x} vs {y}", y=1.02)
    plt.show()
```
