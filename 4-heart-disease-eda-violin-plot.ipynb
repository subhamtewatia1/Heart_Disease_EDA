```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings


warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
df = pd.read_csv("/kaggle/input/heart-data/heart_data.csv")
```
